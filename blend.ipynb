```python
import numpy
import PIL
from PIL import Image
from PIL import ImageOps
import matplotlib.pyplot as plt
```

```python
def resize(im1, im2):
    return ImageOps.fit(im2, (im1.width, im1.height), Image.ANTIALIAS)
```

```python
def resizeBlend(im1, im2, alpha):
    im2 = resize(im1, im2)
    return Image.blend(im1, im2, alpha)
```

```python
im1 = Image.open('vamp2.png')
im2 = Image.open('water2.png')
imshow = resizeBlend(im1, im2, 0.5)
plt.imshow(imshow)
```
